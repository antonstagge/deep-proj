```python
#Imports
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context()
```

```python
#Find path to loss csv files
ROOT_DIR = os.path.abspath("../")
MODEL_DIR = os.path.join(ROOT_DIR, "logs")
LOSS_DIR = os.path.join(MODEL_DIR, "loss_files")
```

```python
#Convert csv files to single pandas dataframe
loss_data = pd.DataFrame()
for filename in os.listdir(LOSS_DIR):
    if filename.endswith(".csv"):
        csvpath = os.path.join(LOSS_DIR, filename)
        loss_type = filename[13:-4]
        df = pd.read_csv(csvpath)
        loss_data[[loss_type]] = df[['Value']]
        print("loaded" + csvpath)
        continue
    else:
        continue
```

```python
#Show evolution of metrics
sns.set()
fig, ax = plt.subplots(figsize = (15,10))
loss_data.plot(ax = ax)
plt.show()
```
